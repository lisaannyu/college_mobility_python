```python
import pandas as pd
```

# Load Data

```python
# table_2: Baseline Cross-Sectional Estimates by College
# table_10: College-level characteristics
# cz (table_10): cz-level characteristics

# Readmes available on the website
```

```python
file_path_table_2 = 'http://www.equality-of-opportunity.org/data/college/mrc_table2.dta'
```

```python
table_2 = pd.read_stata(file_path_table_2)
```

```python
table_10 = pd.read_csv('mrc_table10.csv')
```

```python
file_path_table_cz = 'http://www.equality-of-opportunity.org/data/health/health_ineq_online_table_10.dta'
```

```python
table_cz = pd.read_stata(file_path_table_cz)
```

# Merge Data

```python
# Get list of columns in both dataframes: want to remove one copy so there are no duplicated columns

cols_intersection = table_2.columns.intersection(table_10.columns)\
.drop(['super_opeid']) # Need super_opeid to be the key when joining data frames
```

```python
cols_intersection
```

```python
table_preds = table_2.merge(table_10.drop(cols_intersection, axis = 1, inplace = False), on = 'super_opeid')
```

```python
table_cz = table_cz[table_cz.columns.drop(list(table_cz.filter(regex = '(czname|fips|statename|stateabbrv|smoke|obese|exercise|mort|med_prev|primcare|diab|mammogram|amb_disch)')))]
```

```python
table_preds = table_preds.merge(table_cz, on = 'cz')
```
